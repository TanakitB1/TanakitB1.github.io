### Example 3.6 shows an example of SVM classification for breast cancer

```python
#Example 3.6 Python SVM Breast Cancer Classifications
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

cancer = load_breast_cancer()
X = cancer.data # All of the features
y = cancer.target # All of the labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =
0.2, random_state = 20)

clf = SVC()
clf.fit(X_train, y_train)

#Prediction
y_predict = clf.predict(X_test)

#Print Confusion Matrix and Classification Report
from sklearn.metrics import classification_report, confusion_matrix
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
 columns=['predicted_cancer','predicted_healthy'])
print(confusion)
print(classification_report(y_test, y_predict))
```

### แบบฝึกหัด 3.4
ปรับเปลี่ยนโปรแกรม Python จากตัวอย่างที่ 3.6 ให้สามารถพล็อตฮิสโตแกรมของรัศมี, ขนาด, พื้นผิว, และความเรียบ
 (radius,size, texture, and smoothness)
ของจุดข้อมูลทั้งหมด

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

cancer = load_breast_cancer()
X = cancer.data  # All of the features
y = cancer.target  # All of the labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

clf = SVC()
clf.fit(X_train, y_train)

# Prediction
y_predict = clf.predict(X_test)

# Print Confusion Matrix and Classification Report
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1, 0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'], columns=['predicted_cancer', 'predicted_healthy'])
print(confusion)
print(classification_report(y_test, y_predict))

# Plot histograms for radius, texture, area, and smoothness
features = ["mean radius", "mean texture", "mean area", "mean smoothness"]
feature_indices = [cancer.feature_names.tolist().index(f) for f in features]

plt.figure(figsize=(12, 8))
for i, idx in enumerate(feature_indices):
    plt.subplot(2, 2, i + 1)
    plt.hist(X[:, idx], bins=30, alpha=0.7, color='blue', edgecolor='black')
    plt.title(features[i])
    plt.xlabel(features[i])
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import pickle
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

# Load dataset
X, y = load_iris(return_X_y=True)

# Train Naive Bayes model
clf = GaussianNB()
clf.fit(X, y)

# Save model to file
model_filename = "naive_bayes_iris.pkl"
with open(model_filename, "wb") as model_file:
    pickle.dump(clf, model_file)

# Load model from file
with open(model_filename, "rb") as model_file:
    loaded_clf = pickle.load(model_file)

# Predict using the loaded model
p = loaded_clf.predict([[5.0, 3.4, 1.5, 0.4]])
print(p)
```

### แบบฝึกหัด 3.9

Citations:
[1] https://scikit-learn.org

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X, y = load_diabetes(return_X_y=True)

y = (y > y.mean()).astype(int)  # แปลงให้เป็นค่าที่ยิ่งใหญ่กว่าค่าเฉลี่ยเป็น 1, อื่นๆ เป็น 0

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

clf = RandomForestClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print("Total points: %d Correctly labeled points: %d" % (y_test.shape[0], (y_test == y_pred).sum()))
```

###Example 3.14 compares different classification models.

```python
#Example 3.14 Comparison of Different Classification
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
names = [ "SVM", "Naive Bayes", "LDA",
 "QDA", "Decision Tree", "Random Forest",
 "Nearest Neighbors", "Neural Networks"]
```

```python
classifiers = [
 SVC(),
 GaussianNB(),
 LinearDiscriminantAnalysis(),
 QuadraticDiscriminantAnalysis(),
 DecisionTreeClassifier(),
 RandomForestClassifier(),
 KNeighborsClassifier(),
 MLPClassifier(alpha=1, max_iter=1000)]
```

```python
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
for name, clf in zip(names, classifiers):
 clf.fit(X_train, y_train)
 score = clf.score(X_test, y_test)
 print(name +": " + str(score))
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

names = ["SVM", "Naive Bayes", "LDA", "QDA", "Decision Tree", "Random Forest",
         "Nearest Neighbors", "Neural Networks"]

classifiers = [
    SVC(),
    GaussianNB(),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    KNeighborsClassifier(),
    MLPClassifier(alpha=1, max_iter=1000)
]

X, y = load_diabetes(return_X_y=True)

y = (y > y.mean()).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

for name, clf in zip(names, classifiers):
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print(f"{name}: {score}")
```
